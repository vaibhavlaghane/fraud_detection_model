```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
datasource  = pd.read_csv(  'creditcard_2023_3.csv') 
datasource
```

```python
About Dataset
Description: This dataset contains credit card transactions made by European cardholders in the year 2023. 
    It comprises over 550,000 records, and the data has been anonymized to protect the cardholders' 
    identities. The primary objective of this dataset is to facilitate the development of
    fraud detection algorithms and models to identify potentially fraudulent transactions.

Key Features:

id: Unique identifier for each transaction
V1-V28: Anonymized features representing various transaction attributes (e.g., time, location, etc.)
Amount: The transaction amount
Class: Binary label indicating whether the transaction is fraudulent (1) or not (0)

Data Source: The dataset was collected from credit card transactions made by European cardholders in 2023,
    with sensitive information removed to ensure privacy and compliance with ethical guidelines.
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
 
X = datasource.drop(['Class'],axis=1)
y = datasource[ 'Class']
 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 32, test_size=.3)

X_train#, 
X_test#, 
y_train#, 
y_test

# Answer check
print(  X_train.shape, X_test.shape)
print(X_train.head())
X_train.columns
```

```python
# Cols to Plot 
columns_to_plot =['V1', 'V2', 'V3', 'V4','Class']

# Data Columns
data_to_plot = datasource[columns_to_plot]

# Create a dictionary to map colors to unique values of the 'Quality' column
# Q_colors = { 0 : palette[0], 1 : palette[1], 'other': 'gray'}  

# Creating the pairplot with the specified palette for categorical variables
sns.pairplot(data_to_plot, hue='Class' )
plt.show()
```

```python
# Num _COLS 
NUM_COLS_F =[col for col in datasource.columns if datasource[col].dtype == 'float']

# Define the number of rows and columns for subplots
num_rows = 5  # 4 rows
num_cols = 5  # 4 columns

# Create subplots with appropriate titles
fig, axes = plt.subplots(num_rows, num_cols, figsize=(25, 17))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each numerical column and create a box plot
for i, col in enumerate(NUM_COLS_F[:num_rows * num_cols]):
    sns.boxplot(x=datasource[col], ax=axes[i] )
    axes[i].set_title(col)

# Hide empty subplots
for i in range(len(NUM_COLS_F), num_rows * num_cols):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns
N_d = datasource.select_dtypes(include='number')

# Compute the correlation matrix
correlation_matrix = N_d.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(25, 15))
sns.heatmap(correlation_matrix, annot=True , fmt=".1f", linewidths=0.5)
plt.title('Correlation Plot', fontsize=22)  
plt.tight_layout()  
plt.show()
```

```python
knn_pipe = Pipeline([('scale', StandardScaler()), ('knn', KNeighborsClassifier())])
params = {'knn__n_neighbors': list(range(1, 22, 2))}
knn_grid = GridSearchCV(knn_pipe, param_grid=params)
knn_grid.fit(X_train, y_train)
best_k = list(knn_grid.best_params_.values())[0]
best_acc = knn_grid.score(X_test, y_test)
 
print(best_acc)
print(best_k)
```

```python
params = {'min_impurity_decrease': [0.01, 0.02, 0.03, 0.05],
         'max_depth': [2, 5, 10],
         'min_samples_split': [0.1, 0.2, 0.05]}

grid = GridSearchCV(DecisionTreeClassifier(random_state = 42), param_grid=params).fit(X_train, y_train)
grid_train_acc = grid.score(X_train, y_train)
grid_test_acc = grid.score(X_test, y_test)
best_params = grid.best_params_
 
print(f'Training Accuracy: {grid_train_acc: .2f}')
print(f'Test Accuracy: {grid_test_acc: .2f}')
print(f'Best parameters of tree: {best_params}')
```
